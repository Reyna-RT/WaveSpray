```python
fig,ax = plt.subplots(1,3,figsize=(12,4),sharex=True,sharey=True)

x = np.linspace(0,12,100)
colors = ['k','b','r']
ak = [0.47,0.57,0.66]
mark = ['ws','wo','w^']
for i in range(3):
    mean = w0[w0['A']==amplitude[i]]['size'].mean()
    dist = w0[w0['A']==amplitude[i]]['size']
    variance = np.var(dist/mean)
    kde = st.gaussian_kde(dist/mean)
    ax[0].plot(x,kde(x),mark[i],markeredgecolor='k',markeredgewidth=0.5,markersize=3,
               label='ak = %.2f'%ak[i])
    beta = 1./variance
    gamma = st.gamma.pdf(x,beta,0,variance)
    print(variance)
    ax[0].plot(x,gamma,colors[i],linewidth=1)
ax[0].set_xscale('log')
ax[0].set_yscale('log')
ax[0].legend(fontsize=8)
ax[0].set_ylim(1e-4,1.5)
ax[0].set_xlabel('')
ax[0].set_ylabel('PDF')

#wind=5.2
for i in range(3):
    mean = w5[w5['A']==amplitude[i]]['size'].mean()
    dist = w5[w5['A']==amplitude[i]]['size']
    variance = np.var(dist/mean)
    kde = st.gaussian_kde(dist/mean)
    ax[1].plot(x,kde(x),mark[i],markeredgecolor='k',markeredgewidth=0.5,markersize=3)
    beta = 1./variance
    gamma = st.gamma.pdf(x,beta,0,variance)
    ax[1].plot(x,gamma,colors[i],linewidth=1)
    ax[1].set_xscale('log')
    ax[1].set_yscale('log')
    ax[1].set_ylim(1e-4,1.5)
    ax[1].set_xlabel('')
    ax[1].set_ylabel('PDF')
    
#wind=6.2
for i in range(3):
    mean = w3[w3['A']==amplitude[i]]['size'].mean()
    dist = w3[w3['A']==amplitude[i]]['size']
    variance = np.var(dist/mean)
    kde = st.gaussian_kde(dist/mean)
    ax[2].plot(x,kde(x),mark[i],markeredgecolor='k',markeredgewidth=0.5,markersize=3)
    beta = 1./variance
    gamma = st.gamma.pdf(x,beta,0,variance)
    ax[2].plot(x,gamma,colors[i],linewidth=1,label=r'$\Gamma(x;n)$')
    ax[2].set_xscale('log')
    ax[2].set_yscale('log')
    ax[2].legend(fontsize=8)
    ax[2].set_ylim(1e-4,1.25)

ax[0].set_xlim(0.20,12)
ax[1].set_xlim(0.20,12)
ax[2].set_xlim(0.20,12)
ax[0].set_xlabel(r'$D_{e}/\overline{D_e}$')
ax[1].set_xlabel(r'$D_{e}/\overline{D_e}$')
ax[2].set_xlabel(r'$D_{e}/\overline{D_e}$')
ax[2].set_ylabel('PDF')
ax[0].set_title(r'$U_{max} = 0 m/s$')
ax[1].set_title(r'$U_{max} = 5.2 m/s$')
ax[2].set_title(r'$U_{max} = 6.2 m/s$')
fig.savefig('size-theory-normalized.png',dpi=300,bbox_inches='tight')
```

# Speed Distribution

```python
from scipy.optimize import curve_fit
def Chi3(x,loc,scale):
    from scipy.special import gamma
    y = (x - loc) / scale
    top = y**2 * np.exp(-0.5*y**2)
    bottom = np.sqrt(2) * gamma(1.5)
    return top/(bottom*scale)
```

```python
fig, ax = plt.subplots(1, 3, sharey=True,figsize=(10,3))
x = np.linspace(0,12,50)
for i in range(3):
    speed = np.linalg.norm(np.asarray([w0[w0['A']==amplitude[i]]['u'],
                                       w0[w0['A']==amplitude[i]]['v'],
                                       w0[w0['A']==amplitude[i]]['w']]),axis=0)
    kde = st.gaussian_kde(speed)
    mean = np.nanmax(kde(x))
    print(mean,mean**2 )
    popt3,pcov3 = curve_fit(Chi3,x,kde(x))
    print(popt3)
    y3 = Chi3(x,*popt3)
    ax[i].plot(x,kde(x),mark[i],markeredgecolor='k',markeredgewidth=0.5,markersize=3,
               label='ak = %.2f'%ak[i])
    ax[i].plot(x,y3,'k',linewidth=1,label=r'$M-B(\alpha=$%.2f)'%popt3[1])
    ax[i].legend(fontsize=8)
    ax[i].set_xlim(0,12)
    ax[i].set_ylim(0,0.35)
    ax[i].set_ylabel('PDF')
    ax[i].set_xlabel(r'$|\vec{u}|$ [m/s]')
    ax[i].set_title(r'$U_{max} = 0 m/s$')
#fig.savefig('speed-nowind-dist.png',dpi=300,bbox_inches='tight')
```
